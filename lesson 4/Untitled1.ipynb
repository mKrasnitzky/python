
תרגיל 3: גרף פיזור

    מטרה: לחקור את הקשר בין זמן לימוד לציונים במבחן.
    נתונים: שעות לימוד וציונים של תלמידים.

```python
houres = [2,5,3,4,6,1,5,3]
mark = [95,88,100,70,65,92,99,96]
colors = np.random.rand(8) # 50 צבעים אקראיים
sizes = 1000 * np.random.rand(8) # 50 גדלים אקראיים לנקודות
```

```python
plt.scatter(houres, mark, c=colors, s=sizes, alpha=0.5, cmap='viridis')
plt.colorbar()  # הצגת סרגל הצבעים
plt.title('Scatter Plot Example')
plt.xlabel('houres')
plt.ylabel('mark')
```



    מטרה: לנתח את התפלגות גילאים בקבוצת נבדקים.
    נתונים: גילאים של אנשים בקבוצת מחקר.

```python
age = [50,3,65,45,12,15,18,32,25,24,28]
```

```python
plt.hist(age, bins=[20, 30, 40, 50, 60, 70], edgecolor='black')

plt.title('Age Distribution')
plt.xlabel('Ages')
plt.ylabel('Number of People')
```
